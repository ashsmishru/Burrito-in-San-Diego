#  Predicting overall satisfaction from burrito dimensions

knowing how each dimension will help in overall rating

```python
# Define predictors of the model
features = ['Hunger','Tortilla','Temp','Meat','Fillings','Meat:filling','Uniformity','Salsa','Wrap','Cost']
import statsmodels.api as sm
# here in the below if burritos[] is not used instead if oly np.hstack is used then only overall is added to features as we are using the stored data name the overall rating with all the columns will be included
burritos_new = burritos[np.hstack((features,'overall'))].dropna()
# Run model and display results
X = sm.add_constant(burritos_new[features])
y = burritos_new['overall']

burrito_glm = sm.GLM(y,X)
res = burrito_glm.fit()
newidx = np.argsort(-res.params.values)
temp = np.arange(len(newidx))
newidx = np.delete(newidx,temp[newidx==0])

plt.figure(figsize=(12,4))
plt.bar(np.arange(len(newidx)), res.params[newidx].values, color='.5',yerr=res.bse[newidx].values)
plt.xticks(np.arange(len(newidx)),res.bse[newidx].keys(), size=12)
plt.yticks(np.arange(0,.6,.1), size=12)
plt.ylim((0,.5))
plt.xlabel('Burrito feature', size=20)
plt.ylabel('Model coefficient', size=20)
```

# Principal component analysis

Principal component analysis (PCA) is a technique used for identification of a smaller number of uncorrelated variables known as principal components from a larger set of data. Analysis involves techniques to help produce a smaller number of linear combinations on variables so that the reduced variables account for and explain most the variance in correlation matrix pattern.

```python
from sklearn.decomposition import PCA
# Determine features to use in computing prinicple components
features = burritos[['Volume','Tortilla','Temp','Meat','Fillings','Meat:filling',
            'Uniformity','Salsa','Synergy','Wrap']]

# Replace NaN with mean value (mainly applies to ratings missing Volume measurement)
features = features.fillna(features.mean())

# Normalize each dimension
features = (features - features.mean()) / features.std()

# Apply PCA
pca = PCA(n_components=2)
pca.fit(features)

# Display principal components
pcinfo = {}
pcinfo['Burrito dimension'] = features.keys()
pcinfo['Component 1 weight'] = pca.components_[0]
pcinfo['Component 2 weight'] = pca.components_[1]
print(pd.DataFrame.from_dict(pcinfo))
print('\nExplained variance from first 2 PCs:', pca.explained_variance_ratio_)
```

# Difference between across-restaurant variance and within-restaurant variance

here let us see how the consistancy of the burrito dimension is constant within the shops, across the shop how is the consistency of the burito quality and within with around the shops

```python
feats = ['Tortilla','Temp','Meat','Fillings','Meat:filling','Uniformity','Salsa','Synergy','Wrap','overall', 'Hunger']

# Compute variance within restaurants
var_within_rest = burritos.groupby('Location')[feats].var().mean()
var_across_rest = burritos.groupby('Location')[feats].mean().var()
var_within_by_across = var_within_rest / var_across_rest

df_var = pd.DataFrame({'within shop': var_within_rest,
                       'across shop': var_across_rest, 
                       'within : across': var_within_by_across})
df_var = df_var.stack().reset_index().rename(columns={'level_0':'Feature', 'level_1': 'type', 0: 'Variance'})

plt.figure(figsize=(10,5))
sns.barplot(x='type', y='Variance', hue='Feature', data=df_var, palette='winter',
            hue_order=var_within_by_across.sort_values().index,
            order = ['within shop', 'across shop', 'within : across'])
plt.gca().legend(title='Location', fontsize=15, loc='center left', bbox_to_anchor=(1, 0.5))
plt.setp(plt.gca().get_legend().get_title(), fontsize='15')
plt.xlabel('Variance type', size=15)
plt.ylabel('Normalized Variance', size=15)
plt.xticks(size=15)
plt.yticks(np.arange(0, 2.5, .5), size=15)
plt.ylim((0,2))

feats = ['Volume', 'Cost']
```

# Taking the maximum count of the rating of each dimension from top best to last best tacoshops

```python
import pandas as pd
import matplotlib.pyplot as plt

burritos=pd.read_csv('burritos_01022018.csv')
burritos = burritos.fillna(burritos.mean())
```

```python
x = burritos.groupby('Location').size().reset_index()
temp = x.sort_values(ascending=False, by=0)
temp = temp[:15]
y = pd.DataFrame()
for name in temp['Location']:
    y = y.append(burritos[(burritos['Location']==name)])
```

```python
#Here i am plotting the graph of the total of maximum rating given to the dimension at each restaurant

Volume=[]

cb=burritos[burritos['Location']=='California Burritos']['Volume']
pd.value_counts(cb)
cb=max(pd.value_counts(cb))
cb

ts=burritos[burritos['Location']=='Taco Stand']['Volume']
pd.value_counts(ts)
ts=max(pd.value_counts(ts))
ts

llnp=burritos[burritos['Location']=='Lucha Libre North Park']['Volume']
pd.value_counts(llnp)
llnp=max(pd.value_counts(llnp))
llnp

lpmf=burritos[burritos['Location']=='Los Primos Mexican Food']['Volume']
pd.value_counts(lpmf)
lpmf=max(pd.value_counts(lpmf))
lpmf

Volume.append([cb,ts,llnp,lpmf])

Volume=[7, 7, 4, 5]
Location =['California Burritos','Taco Stand','Lucha Libre North Park','Los Primos Mexican Food']
plt.pie(Volume,labels=Location,radius=2,colors=['Red','Blue','Green','yellow'],autopct='%.1f%%')
print('Variation Of volume ratings at top best to last best restaurants')
```
